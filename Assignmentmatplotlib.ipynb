## **Assignment:
Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
Then create a horizontal bar chart showing total runs scored in their debut 5 years.
**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = ['year1', 'year2', 'year3' 'year4', 'year5']

sachin = [900, 1100, 1000, 1200, 1300]
kohli = [800, 950, 1100, 1150, 1250]
sehwag = [850, 900, 950, 1000, 1050]
```

```python
x = np.arange(len(sachin))
width = 0.25

plt.bar(x - width, sachin, width, label="Sachin")
plt.bar(x, kohli, width, label="Kohli")
plt.bar(x + width, sehwag, width, label="Sehwag")
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = ['Year1', 'Year2', 'Year3', 'Year4', 'Year5']

sachin = [900, 1100, 1000, 1200, 1300]
kohli = [800, 950, 1100, 1150, 1250]
sehwag = [850, 900, 950, 1000, 1050]

x = np.arange(len(years))
width = 0.25

plt.bar(x - width, sachin, width, label="Sachin")
plt.bar(x, kohli, width, label="Kohli")
plt.bar(x + width, sehwag, width, label="Sehwag")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs Comparison (Bar Chart)")

plt.xticks(x, years)
plt.legend()

plt.show()


# Horizontal bar chart
players = ["Sachin", "Kohli", "Sehwag"]

total_runs = [
    sum(sachin),
    sum(kohli),
    sum(sehwag)
]

plt.barh(players, total_runs)

plt.xlabel("Total Runs")
plt.title("Total Runs in First 5 Years")

plt.show()
```

# **Assignment:
Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same
using a horizontal bar chart and observe which is easier to understand.
**

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Android', 'ios', 'Others']
sizes = [25, 35, 40]
```

```python
plt.figure()

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Mobile OS Market Share (Pie Chart)")

plt.show()
```

```python
plt.figure()

plt.barh(labels, sizes)

plt.title("Mobile OS Market Share (Bar Chart)")
plt.xlabel("Market Share (%)")

plt.show()
```
